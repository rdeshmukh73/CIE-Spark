```python
from dotenv import load_dotenv
import os
from google import genai
from google.genai import types
import time
load_dotenv()
```

```python
label = 'identify skin conditions'
words = ['diagnosis', 'decide', 'country', 'person', 'diagnostic', 'skin', 'medicines', 'leading', 'emergencies', 'hospitals']
docs_cleaned = ['There is no standardization of diagnostic machines across diagnostic centres, leading to inconsistent test results and, in some cases, false or delayed diagnosis.',
                'Medical emergencies don‚Äôt wait for doctors. When someone suddenly faints or collapses at home whether it‚Äôs an elderly person, a middle-aged parent, or even a student families often panic and have no idea what to do beyond calling an ambulance. These lost minutes  without even basic first aid can decide whether the person survives or suffers lasting harm.',
                'The core problem observed is the lack of accessible, reliable, and immediate preliminary diagnosis for common skin anomalies, leading to anxiety and unreliable self-diagnosis through non-specialist online sources. This project aims to provide a quick, AI-based visual analysis to help people identify skin conditions and decide if professional care is needed.'
                ]
```

```python
prompt = f"""
You are helping students learn entrepreneurship through problem-based learning.
Your task is to create ONE descriptive problem statement that synthesizes several related issues or topics.
Theme: {label}
Representative words: {words}
Representative problem statements:
- {"\n- ".join(docs_cleaned)}

Guidelines:
1. Write a descriptive problem statement of at least 3–5 sentences.
2. Explain the core issue clearly and why it matters.
3. Identify the primary people affected (stakeholders).
4. Include realistic boundary conditions (affordability, technology, social or infrastructural limitations).
5. Frame it as an open challenge that students can apply innovation and entrepreneurship frameworks to (e.g., NABC, DFV, LeanCanvas or BMC).


Output:
A single well-written problem statement.
Ensure that the Output Problem Statement does not have profanity, slang or bad words/language.
Since the problem statements are going to be used by Students there must not be any words that may be ok in a corporate setting but not appropriate for a university setting.
Nothing related to gender, region, religion, language specific that is inappropriate must be part of the problem statements.
Do not add any new keywords that are not in the originally given Problem Statement text.
Ensure that the problem statements are specific to the given input set of problems and not vague.
Ensure that there are no strong languages or finger pointing words and phrases used towards a specific role or person but instead the problem statement must be generic like say Institution or University or Company or Government without being critical of the university or company or institution.
"""
```

```python
client = genai.Client(api_key = os.getenv('GOOGLE_API_KEY'))

# Gemini models being considered
models = ['gemini-2.5-pro', 'gemini-2.5-flash', 'gemini-2.5-flash-preview-09-2025', 'gemini-2.5-flash-lite', 'gemini-2.5-flash-lite-preview-09-2025', 'gemini-2.0-flash-lite']
# models = ['gemini-2.5-pro']

for m in models :
    print(m.upper())
    # Count input tokens
    input_tokens = client.models.count_tokens(
        model=m, contents=prompt
    )
    print("Input Info")
    print(f'\t Total tokens : {input_tokens.total_tokens}\n\t Cached tokens : {input_tokens.cached_content_token_count}\n')

    # Get response from the model for this prompt
    response = client.models.generate_content(
        model = m,
        contents = prompt
    )
    print("Output Info")
    print(f"Metadata:\t ")
    for i in response.usage_metadata :
        if 'count' in i[0] :
            print(f'{i[0]} : {i[1]}')
    
    print("Response : \n\t", response.text)
    time.sleep(5)
```
